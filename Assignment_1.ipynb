```python
# Create a bill calculator where the user enters the number of units consumed. The rate is
# ₹5 per unit for the first 100 units, ₹7 for the next 100, and ₹10 for anything above 200
# units. Calculate and display the total bill.



units_consumed = int(input("Enter units consumed: "))

#fix rate for units
rate_first_100 = 5
rate_next_100 = 7
rate_above_200 = 10

# initilaise charge value to 0
first_100_charge = 0
next_100_charge = 0
above_200_charge = 0

if units_consumed <= 100:
    first_100_charge = units_consumed * rate_first_100
elif units_consumed <= 200:
    first_100_charge = 100 * rate_first_100
    next_100_charge = (units_consumed - 100) * rate_next_100
else:
    first_100_charge = 100 * rate_first_100
    next_100_charge = 100 * rate_next_100
    above_200_charge = (units_consumed - 200) * rate_above_200

# Total bill
bill_amount = first_100_charge + next_100_charge + above_200_charge

# Output
print(f"Bill amount = {bill_amount}")
```

```python
# Write a Python program to convert a given temperature value between Celsius and
# Fahrenheit. Accept input as a number and a unit indicator ('C' or 'F').

#  Reads Input
temp_value = float(input("Enter temperature value: "))
unit = input("Enter unit (C/F): ").strip().upper()

# logic
if unit == "C":
    converted_temp = (temp_value * 9/5) + 32
    print(f"Temperature in Fahrenheit: {converted_temp:.1f}°F")
elif unit == "F":
    converted_temp = (temp_value - 32) * 5/9
    print(f"Temperature in Celsius: {converted_temp:.1f}°C")
else:
    print("Invalid unit. Please enter 'C' for Celsius or 'F' for Fahrenheit.")
```

```python

# Write a program that takes the recharge date and today’s date as input and checks
# whether the 28-day recharge plan is still valid or expired

from datetime import datetime

recharge_date = input("Enter recharge date (YYYY-MM-DD): ")
today_date = input("Enter today's date (YYYY-MM-DD): ")
#using datetime function to extract date from string
rd = datetime.strptime(recharge_date, "%Y-%m-%d")
td = datetime.strptime(today_date, "%Y-%m-%d")

days_passed = (td - rd).days

if days_passed <= 28:
    print("Plan is still valid.")
else:
    print("Plan expired. Recharge again.")
```

```python
# A list contains the names of passengers in order of booking. A cancellation occurs; remove
# the name from the list and shift the rest. Also, insert a new booking at a specified position.

# inserting values into list
booked = input("Enter booked passenger names separated by commas: ").split(',')

# Removing spaces
for i in range(len(booked)):
    booked[i] = booked[i].strip()

# Take cancellation and update the change
cancel = input("Enter the name to cancel: ").strip()
new = input("Enter the new passenger name: ").strip()
position = int(input("Enter the position to insert (0 for first position): "))

# Remove cancelled passenger
if cancel in booked:
    booked.remove(cancel)
else:
    print(f"{cancel} not found in the booking list.")

# Insert the  new passenger
booked.insert(position, new)

#output the  updated list
print("Updated booking list:", booked)
```

```python

# Write a program that takes a number as input and prints the sum of all even numbers
# from 1 to that number.
# Take number from user
n = int(input("Enter a number: "))

# Initialize sum
even_sum = 0

# Loop through numbers to  add even values
for i in range(1, n + 1):
    if i % 2 == 0:
        even_sum += i

# output the value
print("Sum of even numbers from 1 to", n, "is:", even_sum)
```

```python
# Take sentence as input
sentence = input("Enter a sentence: ")

# Split into words and convert to lowercase
words = sentence.lower().split()

# Create empty dictionary for frequency value
word_freq = {}

# Count number of  occurrences
for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1


for word, freq in word_freq.items():
    print(f"{word}: {freq}")

```

```python
# Write a function that checks whether a number is prime. Call the function with user input.
# Function to check prime
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num/2) + 1):
        if num % i == 0:
            return False
    return True

# Take user input
number = int(input("Enter a number: "))

# Call function and print the  result
if is_prime(number):
    print(f"{number} is a prime number")
else:
    print(f"{number} is not a prime number")
```

```python
# Write a function that takes a string and returns the number of vowels in it.
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

# Take input from the  user
string_input = input("Enter a string: ")

# Call  function and display result,no of vowels
print("Number of vowels:", count_vowels(string_input))
```

```python
# Take inputs limit and divisor
N = int(input("Enter range limit N: "))
D = int(input("Enter divisor D: "))

# Counter
count = 0

# Loop through numbers and check if divisible
for i in range(1, N + 1):
    if i % D == 0:
        count += 1

# Output
print(f"Numbers divisible by {D} between 1 and {N}: {count}")
```
